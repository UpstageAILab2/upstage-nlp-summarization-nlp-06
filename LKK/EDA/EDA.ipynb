```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('../data/train.csv')
dev = pd.read_csv('../data/dev.csv')
test = pd.read_csv('../data/test.csv')
```

```python
display(train.shape, dev.shape, test.shape)
```

```python
train_dialogue_length = train.dialogue.apply(lambda x: len(x))
dev_dialogue_length = dev.dialogue.apply(lambda x: len(x))
test_dialogue_length = test.dialogue.apply(lambda x: len(x))

train_summary_length = train.summary.apply(lambda x: len(x))
dev_summary_length = dev.summary.apply(lambda x: len(x))
# test_summary_length = test.summary.apply(lambda x: len(x)) #test는 summary가 없음
```

```python
f, ax = plt.subplots(1,1)
sns.histplot(train_dialogue_length, alpha=0.3, ax=ax, label='train_dialogue')
sns.histplot(dev_dialogue_length, alpha=0.3, ax=ax, label='dev_dialogue')
sns.histplot(test_dialogue_length, alpha=0.3, ax=ax, label='test_dialogue')
ax.legend()
```

```python
f, ax = plt.subplots(1,1)
sns.kdeplot(train_dialogue_length, alpha=0.5, ax=ax, label='train_dialogue')
sns.kdeplot(dev_dialogue_length, alpha=0.5, ax=ax, label='dev_dialogue')
sns.kdeplot(test_dialogue_length, alpha=0.5, ax=ax, label='test_dialogue')
ax.legend()
```

```python
f, ax = plt.subplots(1,1)
sns.histplot(train_summary_length, alpha=0.3, ax=ax, label='train_summary')
sns.histplot(dev_summary_length, alpha=0.3, ax=ax, label='dev_summary')
ax.legend()
```

```python
f, ax = plt.subplots(1,1)
sns.kdeplot(train_summary_length, alpha=0.5, ax=ax, label='train_summary')
sns.kdeplot(dev_summary_length, alpha=0.5, ax=ax, label='dev_summary')
ax.legend()
```
